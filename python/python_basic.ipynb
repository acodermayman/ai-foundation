```python
#function derivative y 2*x^2 + x
def fx(x):
    return x**2 + x
def derivative(x, delta_x):
    # return (fx(x + delta_x) - fx(x)) / delta_x
    x0 = x
    f0 = fx(x0)
    x1 = x0 + delta_x
    f1 = fx(x1)
    return (f1 - f0) / delta_x
print(derivative(2,0.0001))
```

```python
#If-Else
import math
def function_1(a,b):
    K=10
    if a == b:
        return b*b
    if a == K:
        return math.sqrt(b)
print(function_1(0,4))
print(function_1(10,4))
```

```python
#Alternative to If-Else
import math
def function_1(a,b):
    K=10
    return (a/K)*math.sqrt(b) + ((K-a)/K)*b*b
print(function_1(0,4))
print(function_1(10,4))
```

```python
import math
import numpy as np

def function_3(a,b):
    vector_v = np.array([0,0,0])
    vector_v[a] = 1
    vector_u = ([b*b, math.sqrt(b),b])
    return np.dot(vector_v,vector_u)
print(function_3(0,4))
print(function_3(1,4))
print(function_3(2,4))
```

```python
#Pi Gregory-Leibniz Series
n  = 1000
PI = 0
for i in range(1,n):
    PI += (-1)**(i+1) / (2*i - 1)
PI *= 4
print(PI)
```

```python
#PI Nilakantha Series
n  = 1000
PI = 0
for i in range(n):
    PI += (-1)**i / ((2*i+2)*(2*i+3)*(2*i+4))
PI = 3 + 4*PI 
print(PI)
```

```python
#Quadratic Root - Newton Method
def compute_square_root(a,n):
    results = a/2.0
    for _ in range(n):
        results = (results + a/results) / 2
    return results
print(compute_square_root(9,5))
print(compute_square_root(16,5))
```

```python
#E estimation
def factorial(a):
    for i in range(1,a): 
        a*=i
    return a
E = 1.0
for i in range(1,1000):
    E+=1/factorial(i) 
print(E)
```

```python
#Compute the area of a unit circle
import math

delta_x = 0.01
n = int(2/delta_x)
haft_area = 0.0
x = -1.0

def compute_y(x):
    return math.sqrt(1-x*x)

for _ in range(n):
    y = compute_y(x)
    haft_area += delta_x*y
    x += delta_x

print(2*haft_area)
```

```python
def F1_Score(tp, fp, fn):
    if not isinstance(tp, int):
        print("tp must be int")
    elif not isinstance(fp, int):
        print("fp must be int")
    elif not isinstance(fn, int):
        print("fn must be int")
    elif tp <= 0:
        print("tp must be greater than zero")
    elif fp <= 0:
        print("fp must be greater than zero")
    elif fn <= 0:
        print("fn must be greater than zero")
    else:
        precision = tp / (tp + fp)
        print(f'precision: {precision}')
        recall = tp / (tp + fn)
        print(f'recall: {recall}')
        f1_score = 2 * (precision * recall) / (precision + recall)
        print(f'f1_score: {f1_score}')
    return

F1_Score(5, 4, 3)
```

```python
#Activation Functions
import math

def is_number(x):
    if not isinstance(x,int) and not isinstance(x,float):
        print('x must be a number')
        return False
    return True

def sigmoid(x):
    return 1/(1+math.e**(-x))

def relu(x):
    if x <= 0:
        return 0
    else:
        return x

def ELU(x):
    alpha = 0.01
    if x <= 0:
        return alpha*(math.e**x - 1)
    else:
        return x

def activation_function(x, act_name):
    if not is_number(x):
        return

    valid_names = ["sigmoid", "relu", "ELU"]
    if act_name not in valid_names:
        print("function_user is not supported")
        return

    if act_name == 'sigmoid':
        return sigmoid(x)
    if act_name == 'relu':
        return relu(x)
    if act_name == 'ELU':
        return ELU(x)

act_name = input('Nhap action name: ')
x = int(input('Nhap x: '))
print(activation_function(x,act_name))
```
